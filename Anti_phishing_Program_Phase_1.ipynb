The program checks messages against the built-in phrase list, then classifies each message as SAFE, SUSPICIOUS, or DANGEROUS.

The program uses the following global data structure:

* phishing_phrases — a list of 10 known phishing
phrases (strings)


```python
phishing_phrases = ["verify your account",
"click here to claim",
"you have won",
"update your payment",
"confirm your password",
"your account will be closed",
"free gift card",
"send your otp",
"urgent action required",
"your parcel is on hold"]

community_reports = ["click to claim a prize"]
```

```python
phishing_phrases
```

```python
def check_message(message):
  message_lower = message.lower()

  flagged = []
  match = 0

  for phrase in phishing_phrases:
    if phrase in message_lower:
      flagged.append(phrase)

  if len(flagged) == 0:
    print("SAFE and No phishing signals found.")
  elif len(flagged) == 1:
    print("SUSPICIOUS and Proceed with caution!")
  else:
    print("DANGEROUS and This looks like a phishing message. Delete it!")

  print(f"Number of flags found: {len(flagged)}")

  print("These are the flagged phrases: ")
  for phrase in flagged:
    print(phrase)

  for i in range(len(flagged)):
    print(f"{i+1}: {flagged[i]}")

  for phrase in community_reports:
    message_word = message_lower.split()
    community_word = phrase.split()

  for word in message_word:
    if word in community_word:
      match += 1
  print("number of matches in community report:", match)

  if match >= len(community_word)//2:
    flagged.append("Similar to a community-reported phishing message")
  return flagged

#  for word in message_word:
#    if word in community_word:
#      print(word)
```
